```python
import pandas as pd

csv = pd.read_csv("daily_temperature_1000_cities_1980_2020.csv")
tempdf = pd.DataFrame(csv)
tempdf.head(25)
```

```python
tempdf.index = tempdf.iloc[:,0]
tempdf.drop(columns='Unnamed: 0', inplace=True)
tempdf
```

```python
citytemp = tempdf.T
citytemp
```

```python
citytemp['city_code'] = citytemp.index
cityDF = citytemp.reset_index(drop=True)
```

```python
#set columns that will not pivot
id_vars = cityDF.columns[0:12].to_list()
id_vars.append("city_code")

#pivot on Dates
city_final = cityDF.melt(id_vars=id_vars, var_name='Date', value_name = 'Temp')
city_final
```

### Temperature Change chart

```python
# import plotly.graph_objects as go
import plotly.express as px

#select city to filter on
city = 'Tokyo'

#build chart
df = city_final[city_final['city'] == city]
fig = px.line(df, x="Date", y="Temp", title=f"Temperature Change in {city}")
fig.write_html('city_temp.html', auto_open=True)
```

### Population Change chart

```python
#select city to filter on
city = 'New York'

#build chart
df = city_final[city_final['city'] == city]
fig = px.line(df, x="Date", y="population", title=f"Population Change in {city}")
fig.write_html('city_pop.html', auto_open=True)

#####################################
#no population data in this DF
#####################################
```
